# Data Cleaning & Preprocessing
## 1. Load Dataset & Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

DataSet = pd.read_csv('../data/SCMS_Delivery_History_Dataset.csv', encoding="ISO-8859-1")

DataSet.info()
DataSet.head()
```

## 2. Check for Missing Values

```python
print(DataSet.isnull().sum())
```

## 3. Clean Missing Values & Convert Data Types

```python
DataSet['Shipment Mode'].fillna("Unknown", inplace=True)
DataSet['Dosage'].fillna("Unknown", inplace=True)
    
DataSet['Line Item Insurance (USD)'].fillna(DataSet['Line Item Insurance (USD)'].median(), inplace=True)

date_columns = [
    "PQ First Sent to Client Date",
    "PO Sent to Vendor Date",
    "Scheduled Delivery Date",
    "Delivered to Client Date",
    "Delivery Recorded Date"
]

for col in date_columns:
    DataSet[col] = pd.to_datetime(DataSet[col], errors='coerce')
    
DataSet['Weight (Kilograms)'] = pd.to_numeric(DataSet['Weight (Kilograms)'], errors='coerce')
DataSet['Freight Cost (USD)'] = pd.to_numeric(DataSet['Freight Cost (USD)'], errors='coerce')

print("\nUpdated Data Types:")
print(DataSet.dtypes)
```

```python
unique_counts = DataSet.nunique().sort_values(ascending=False)
print("\nUnique Values per Column:")
print(unique_counts)
```

## 4.Drop Irrelevant or Redundant Columns

```python
columns_to_drop = [
    "ID", "PQ #", "PO / SO #", "ASN/DN #",
    "Item Description", "Line Item Insurance (USD)", "Managed By", "Dosage Form", "Dosage", "Molecule/Test Type", "Project Code", 
]

DataSet.drop(columns=columns_to_drop, inplace=True)

DataSet.head()
```

```python
columns_to_drop1 = [
    "Brand", "First Line Designation", "Unit of Measure (Per Pack)",
    "Pack Price", "Unit Price"
]

DataSet.drop(columns=columns_to_drop1, inplace=True)
DataSet.head()
```

## 5. Create Target & Impute Missing Values w/ Median

```python
DataSet["Delivery Delay (Days)"] = (DataSet["Delivered to Client Date"] - DataSet["Scheduled Delivery Date"]).dt.days

DataSet["Delayed"] = (DataSet["Delivery Delay (Days)"] > 0).astype(int)

DataSet["Weight (Kilograms)"].fillna(DataSet["Weight (Kilograms)"].median(), inplace=True)
DataSet["Freight Cost (USD)"].fillna(DataSet["Freight Cost (USD)"].median(), inplace=True)

DataSet.dropna(subset=["Scheduled Delivery Date", "Delivered to Client Date"], inplace=True)

for col in ["PQ First Sent to Client Date", "PO Sent to Vendor Date"]:
    DataSet[col].fillna(pd.Timestamp("2000-01-01"), inplace=True)
    
print(DataSet.isnull().sum())
```

## 6. Feature Engineering

```python
DataSet["Delivery Quarter"] = DataSet["Scheduled Delivery Date"].dt.quarter
```

```python
DataSet["Freight Cost per KG"] = DataSet["Freight Cost (USD)"] / DataSet["Weight (Kilograms)"]

DataSet.head()
```

```python
features_to_drop = [
    "PQ First Sent to Client Date", "PO Sent to Vendor Date", "Scheduled Delivery Date", "Delivered to Client Date", "Delivery Recorded Date"
]
    
DataSet.drop(columns=features_to_drop, inplace=True)
DataSet.head()
```

## 7. One-Hot Encoding Specified Variables


```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

one_hot_cols = ["Fulfill Via", "Shipment Mode", "Vendor INCO Term", "Country", "Product Group", "Delivery Quarter"]

DataSet = pd.get_dummies(DataSet, columns=one_hot_cols, drop_first=False)
DataSet.info()
```

```python
 DataSet.drop(columns=["Sub Classification", "Delivery Delay (Days)", "Manufacturing Site", ], inplace=True)
```

## 8. Frequency Encoding for Vendor

```python
vendor_counts = DataSet["Vendor"].value_counts()
DataSet["Vendor_encoded"] = DataSet["Vendor"].map(vendor_counts)

DataSet = DataSet.drop(columns=["Vendor"], errors="ignore")

print(DataSet[["Vendor_encoded"]].head())
```

```python
processed_file_path = "../data/SCMS_Encoded1.csv"
DataSet.to_csv(processed_file_path, index=False)
```

## 9. Identify & Drop Highly Correlated/Redundant Variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import RFE, mutual_info_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler

correlation_matrix = DataSet.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=False)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
correlation_threshold = 0.5
correlated_pairs = []

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            correlated_pairs.append((correlation_matrix.columns[i], correlation_matrix.columns[j], correlation_matrix.iloc[i, j]))

correlated_df = pd.DataFrame(correlated_pairs, columns=["Feature 1", "Feature 2", "Correlation"])
print("Highly Correlated Feature Pairs:")
print(correlated_df)
```
